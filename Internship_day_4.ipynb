```python
#student score analysis
import numpy as np
scores = np.array([85,76,92,99,44,89])

print("Scores:",scores [0:5])
print("Average Score:",np.mean(scores))
print("Highest Score:",np.max(scores))
print("Lowest Score:",np.min(scores))
print("Total Score:",np.sum(scores))
```

```python
#sales array calculation
import numpy as np

sales=np.array([12000,15000,18000,14000,20000])
print("Sales:",sales)
print("Average Sales:",np.mean(sales))
print("Highest Sales:",np.max(sales))
print("Lowest Sales:",np.min(sales))
print("Total Sales:",np.sum(sales))
```
