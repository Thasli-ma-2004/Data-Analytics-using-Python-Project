```python
#pie chart
import matplotlib.pyplot as plt

expenses=[5000,3000,2000]

categories=["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories,autopct="%1.1f%%",colors=["Red","yellow","green"])

plt.title("Expense Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr","May","Jun"]

sales=[10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(11,4))

#Line Chart

plt.subplot(1,3,1)

plt.plot(months,
         sales,
         marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

#Bar Chart
plt.subplot(1,3,2)
plt.bar(months,
        sales,
        color="orange")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")

#pie chart
plt.subplot(1,3,3)

expenses=[5000,3000,2000]
categories=["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories,autopct="%1.1f%%",colors=["Red","yellow","green"])

plt.title("Expense Distribution")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

marks=[45,50,55,60,65,70,72,75,80,85,90,95]

plt.hist(marks,bins=5,edgecolor="black")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Students Marks Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

hours=[1,2,3,4,5,6,7,8]

marks=[35,40,50,60,65,75,85,95]

plt.scatter(hours,marks)

plt.title("Study Hours vs Marks")

plt.xlabel("Study Hours")

plt.ylabel("Marks")

plt.show()
```
