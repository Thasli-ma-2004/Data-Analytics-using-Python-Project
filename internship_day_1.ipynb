```python
def get_grade(mark):
  if mark>=90:
    return 'S'
  elif mark>=75:
      return 'A'
  else:
       return 'F'

students = {
   "Thaslima": 97,
   "Gayathri": 92,
   "Thaniksha": 80,
   "Divya": 65
}

passed=0
failed=0

for name,mark in students.items():
  grade = get_grade(mark)
  print(name,"-",mark,"-",grade)

  if grade =='F':
    failed +=1
  else:
    passed +=1


print("Passed",passed)
print("Failed",failed)
```
